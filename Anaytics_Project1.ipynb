```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel(r"C:\Users\nt397\Downloads\Mobiles Dataset (2025).xlsx")
df
```

```python
df.isnull().sum()
```

```python
price_columns=["Launched Price (Pakistan)","Launched Price (India)","Launched Price (China)","Launched Price (USA)","Launched Price (Dubai)"] 
price_columns
```

```python
for i in price_columns:
    df[i] = df[i].replace(
        ['Not available', 'N/A', 'null', 'NaN', 'missing'], 
        0
    )
```

```python
df
```

```python
s=df[price_columns]
s
```

```python
for col in price_columns:
    df[col] = df[col].map(lambda x: ''.join(c for c in str(x) if c.isdigit()))
df
```

```python
df.info()
```

```python
df[price_columns]=df[price_columns].astype('int')
df.info()
```

```python
import matplotlib.pyplot as plt
```

```python
for col in price_columns:
    plt.boxplot(df[col].dropna())
    plt.title(f'Boxplot for {col}')
    plt.ylabel('Price')
    plt.show()
```

```python
s=df['Launched Price (Pakistan)']
s
```

```python
q1=np.quantile(s,0.25)
q1
```

```python
q2=np.quantile(s,0.75)
q2
```

```python
h=q2-q1
h
```

```python
upper=q2+(1.5*h)
upper
```

```python
lower=q1-(1.5*h)
lower
```

```python
p=[]
for i in s:
    if(i>lower and i<upper):
        p.append(i)
p
```

```python
plt.boxplot(p)
plt.title(f'Boxplot for Launched Price (Pakistan) ')
plt.show()
```
